```python
class Solution:
    def convert(self, s: str, numRows: int) -> str:
        # 2 s
        if numRows < 2: return s
        # 
        res = ["" for _ in range(numRows)]
        #   
        i, flag = 0, -1
        for c in s:
            res[i] += c
            #  
            if  i == numRows - 1 or i == 0: 
                flag = -flag
            # 
            i += flag
        return "".join(res)
```
