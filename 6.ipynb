思路：  
最后的答案由numRows个字符串拼接而成，就那构造numRows个字符串   
遍历s的同时，先正向的遍历这numRows个字符串，并追加当前字符；当到达N的转折点时，反向遍历字符串，并追加字符；  
  
如图：  
先准备4个空字符串，依次遍历s  
正向：
步骤一：c = P , numRows[0] = c  
步骤二：c = A , numRows[1] = c  
...  
步骤四：c = P , numRows[3] = c
达到转折点，反向：  
步骤一：c = A , numRows[2] = c  
步骤二：c = L , numRows[1] = c   
步骤三：c = I , numRows[0] = c  
达到转折点，正向：
...  
如何控制方向呢？  
维护一个flag 当遇到转折点时，flag会被更新为-flag  
初始为-1 由于刚开始时也是转折点 他会被更新为1 也就是正向  
每当 i == 行号 -1就修改方向，如上例 就是每遍历3个字符就修改方向

```python
class Solution:
    def convert(self, s: str, numRows: int) -> str:
        # 如果小于2行 s就是结果
        if numRows < 2: return s
        # 对每一行都准备一个空串
        res = ["" for _ in range(numRows)]
        # 答案串的索引 和 方向
        i, flag = 0, -1
        for c in s:
            res[i] += c
            # 如果到达转折处 就更新方向
            if  i == numRows - 1 or i == 0: 
                flag = -flag
            # 选择下一个串
            i += flag
        return "".join(res)
```

```python
Solution().convert(s = "PAYPALISHIRING", numRows = 4)
```
