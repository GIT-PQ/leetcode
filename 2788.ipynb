思路：  
遍历每个字符串  
对每个字符串通过操作符进行分割  只保留非空的分割结果  


```python
class Solution:
    def splitWordsBySeparator(self, words, separator: str):
        res = []
        for string in words:
            res += [word for word in string.split(separator) if len(word)]
        return res
```

```python
S = Solution()
S.splitWordsBySeparator(words = ["one.two.three","four.five","six"], separator = ".")
```
