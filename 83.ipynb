```python
from typing import Optional
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head == None:
            return head
        pre = head
        last = head.next
        while last:
            if pre.val == last.val:
                pre.next = last.next
                last=pre.next
            else:
                pre = last
                last = last.next
        return head
```

```python
node4 = ListNode(3)
node3 = ListNode(2,node4)
node2 = ListNode(2,node3)
node1 = ListNode(1,node2)
head = ListNode(1,node1)
print("",end="")
p = head
while p:
    print(p.val,end=" ")
    p = p.next 

S=Solution()
S.deleteDuplicates(head)
print("",end="")
while head:
    print(head.val,end=" ")
    head = head.next 
```
