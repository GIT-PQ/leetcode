思路：散列表  
准备两个散列表，pre和last，分别存储nums[:i+1]内数字:个数，nums[i:]内的数字:个数  
然后动态的更新pre和last，计算(len(pre) - len(last))放到res列表中即可  

```python
from typing import List
class Solution:
    def distinctDifferenceArray(self, nums: List[int]) -> List[int]:
        res,pre,last=[],{},{}
        # 初始化last
        for dig in nums:
            last[dig] = last.get(dig,0)+1
        for i,dig in enumerate(nums):
            pre[dig] = pre.get(dig,0)+1
            last[dig] -= 1
            if last[dig] == 0:
                del last[dig]
            res.append(len(pre) - len(last))
        return res
```

```python
Solution().distinctDifferenceArray([3,2,3,4,2])
```
