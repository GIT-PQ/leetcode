思路：  
所谓最优策略 就是我拿石子价值高的同时，对方损失的价值也最高
那么问题就简化成了每次拿最大的aliceValues[i] + bobValues[i]   (1)即可  
容易想到，将(1)按照值进行排序，然后alice取奇数index对应的元素累加，bob取偶数的index对应的元素累加，最后比较两者谁的价值高  

```python
from typing import List
class Solution:
    def stoneGameVI(self, aliceValues: List[int], bobValues: List[int]) -> int:
        values = [[a+b, a, b] for a, b in zip(aliceValues, bobValues)]
        values.sort(reverse=True)
        aliceSum, bobSum = sum(value[1] for value in values[::2]), sum(value[2] for value in values[1::2])
        if aliceSum > bobSum:
            return 1
        elif aliceSum == bobSum:
            return 0
        else:
            return -1
```

```python
Solution().stoneGameVI(aliceValues = [1,3], bobValues = [2,1])
```
