思路：  
对数组的每个元素进行遍历，对于每个元素如果满足nums[i] - nums[i+1] == (-1)**string_len  (1) string_len代表子数组的长度  
就令子数组长度+1
否则就判断最大子数组长度和当前子数组哪个大 取最大作为最大子数组长度 然后将子数组长度置为1  (2)  

最后再判断一次是否满足(1) 这是因为可能存在这种情况 345 345不满足 当前索引为2 但是45却满足 而(2)将长度置为1后，索引就往后走为3了  
这样会导致错误

```python
from typing import List
class Solution:
    def alternatingSubarray(self, nums: List[int]) -> int:
        string_len = 1
        max_len = -1
        n = len(nums)
        for i in range(n):
            if i+1 < n and nums[i] - nums[i+1] == (-1)**string_len:
                string_len += 1
            else:
                max_len = string_len if string_len > max_len and string_len > 1  else max_len
                string_len = 1
            if i+1 < n and nums[i] - nums[i+1] == (-1)**string_len:
                string_len += 1
        return max_len
```

```python
S = Solution()
S.alternatingSubarray([2,3,4,3,4])
```
