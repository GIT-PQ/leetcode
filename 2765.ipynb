```python
from typing import List
class Solution:
    def alternatingSubarray(self, nums: List[int]) -> int:
        string_len = 1
        max_len = -1
        n = len(nums)
        for i in range(n):
            if i+1 < n and nums[i] - nums[i+1] == (-1)**string_len:
                string_len += 1
            else:
                max_len = string_len if string_len > max_len and string_len > 1  else max_len
                string_len = 1
            if i+1 < n and nums[i] - nums[i+1] == (-1)**string_len:
                string_len += 1
        return max_len
```
