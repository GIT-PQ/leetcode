思路：  
依次交换所有的两个数字，记录下最大的即可

```python
class Solution:
    def maximumSwap(self, num: int) -> int:
        res = num
        s = list(str(num))
        for i in range(len(s)):
            for j in range(i):
                s[i],s[j] = s[j],s[i]
                res = max(res,int(''.join(s)))
                s[i],s[j] = s[j],s[i]
        return res
```

```python
S =Solution()
S.maximumSwap(2736)
```
