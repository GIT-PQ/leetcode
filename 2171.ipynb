思路：  
将数组排序，对于每个元素beans[i]，它前面的袋子全部清空，后面的袋子全部保留beans[i]颗魔法豆  
假设魔法豆总数为x，袋子数量为n，最小数目为min{x-beans[i][n-i]}

```python
class Solution:
    def minimumRemoval(self, beans) -> int:
        n = len(beans)
        x = sum(beans)
        amount = x
        beans = sorted(beans)
        for i,num in enumerate(beans):
            amount = min(amount,(x-num*(n-i)))
        return amount
```

```python
beans = [2,10,3,2]
S = Solution()
S.minimumRemoval(beans)
```
