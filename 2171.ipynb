  
beans[i]beans[i]  
xnmin{x-beans[i][n-i]}

```python
class Solution:
    def minimumRemoval(self, beans) -> int:
        n = len(beans)
        x = sum(beans)
        amount = x
        beans = sorted(beans)
        for i,num in enumerate(beans):
            amount = min(amount,(x-num*(n-i)))
        return amount
```

```python
beans = [2,10,3,2]
S = Solution()
S.minimumRemoval(beans)
```
