思路：  
类似于冒泡的遍历方式，依次遍历每个word，每个word只需要和word后面的word进行比较即可

```python
class Solution:
    def maximumNumberOfStringPairs(self, words) -> int:
        n = len(words)
        amount = 0
        for i in range(n):
            for j in range(i+1,n):
                if words[j][::-1] == words[i]:
                  amount += 1
        return amount
```

```python
words = ["cd","ac","dc","ca","zz"]
S = Solution()
S.maximumNumberOfStringPairs(words)
```

思路2：哈希表  
对words进行一次遍历，如果当前word取反后在hash中，那么amount就加一，最后将当前word放入hash中

```python
class Solution:
    def maximumNumberOfStringPairs(self, words) -> int:
        n = len(words)
        hash = set()
        amount = 0
        for i in range(n):
            if words[i][::-1] in hash:
                amount+=1
            hash.add(words[i])
        return amount
```

```python
words = ["cd","ac","dc","ca","zz"]
S = Solution()
S.maximumNumberOfStringPairs(words)
```
