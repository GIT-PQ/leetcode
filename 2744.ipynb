```python
class Solution:
    def maximumNumberOfStringPairs(self, words) -> int:
        n = len(words)
        amount = 0
        for i in range(n):
            for j in range(i+1,n):
                if words[j][::-1] == words[i]:
                  amount += 1
        return amount
```

```python
words = ["cd","ac","dc","ca","zz"]
S = Solution()
S.maximumNumberOfStringPairs(words)
```

```python
class Solution:
    def maximumNumberOfStringPairs(self, words) -> int:
        n = len(words)
        hash = set()
        amount = 0
        for i in range(n):
            if words[i][::-1] in hash:
                amount+=1
            hash.add(words[i])
        return amount
```

```python
words = ["cd","ac","dc","ca","zz"]
S = Solution()
S.maximumNumberOfStringPairs(words)
```
