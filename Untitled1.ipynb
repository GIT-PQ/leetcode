```python
import nibabel as nib
import matplotlib.pyplot as plt

# 用您的 NII 文件路径替换下面的路径
nii_file_path = "D:\data\cv\RibFrac1-image.nii"

# 使用 nibabel 打开 NII 文件
nii_img = nib.load(nii_file_path)

# 获取图像数据
data = nii_img.get_fdata()

# 显示图像的某一层（这里是第一层）
plt.imshow(data[:, :, 4], cmap="gray")
plt.title("NII Image - Slice 0")
plt.show()
```
