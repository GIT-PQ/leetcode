思路：通过一个字典来存储 word：v  
如果word在words1中出现就+1  
在words2中出现就+10  
最后遍历字典，确定v==11即可代表word在两个数组中都只出现了一次  

```python
class Solution:
    def countWords(self, words1, words2):
        dic = {}
        amount = 0
        for word in words1:
            dic[word] = dic.get(word,0)+1
        for word in words2:
            dic[word] = dic.get(word,0)+10
        for k,v in dic.items():
            if v == 11:
                amount+=1
        return amount
```

```python
S = Solution()
words1 = ["leetcode","is","amazing","as","is"]
words2 = ["amazing","leetcode","is"]
S.countWords(words1,words2)
```
