思路：  
对每个字符极其后面的字符进行遍历，如果存在重复就结束内层循环，移动到下一个字符
将之前已经便利的字符放到集合中

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        if s == '':
            return 0
        n = len(s)
        max_len = 1;
        hash = set()
        for i,c in enumerate(s):
            pre = c;
            now_len = 1
            hash.add(c)
            for j in range(i+1,n):
                if s[j] in hash:
                    break
                else:
                    now_len += 1
                    max_len = max_len if max_len > now_len else now_len
                    hash.add(s[j])
            hash.clear()
        return max_len
```

```python
S=Solution()
S.lengthOfLongestSubstring('abcabcbb')
```
